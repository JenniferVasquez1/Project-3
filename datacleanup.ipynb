### Extract the dataset files for:
- AQI.csv
- Natural_Gas_Consumption_Data.csv

```python
aqi_df = pd.read_csv('static/Resources/aqi.csv')
aqi_df.head()
```

```python
aqi_df.shape
```

```python
aqi_df['Date'] = pd.to_datetime(aqi_df['Date'])
aqi_df['YearMonth'] = aqi_df['Date'].dt.to_period('M')

aqi_df.head(20)
```

```python
#aggregating air_quality data by state and month and calculating the average AQI for all cities within a state
state_monthly_avg_aqi = aqi_df.groupby(['state_id','YearMonth']).agg({'AQI':'mean'}).reset_index()

state_monthly_avg_aqi['AQI'] = state_monthly_avg_aqi['AQI'].round(2)

state_monthly_avg_aqi.head(10)
```

```python
#adding category column based on the AQI average value
bins = [0, 50, 100, 200, 350, 600]
labels = ['Good', 'Moderate', 'Unhealthy', 'Very Unhealthy', 'Hazardous']
state_monthly_avg_aqi['category'] = pd.cut(state_monthly_avg_aqi['AQI'], bins=bins, labels=labels, right =False)

state_monthly_avg_aqi.head(10)
```

```python
state_monthly_avg_aqi.category.value_counts()
```

### Data Cleaning 
- Removing unwanted columns
- Reorganizing/Renaming the columns
- Checking for Null values
- Checking the datatypes of our data Frames

```python
aqi_df.columns
```

```python
# Removing the unwanted columns of aqi_df into and storing it into new dataframe
cleaned_aqi_df = aqi_df[['Date','AQI','Category','Defining Parameter','state_id','lat','lng','population','density']]

cleaned_aqi_df.head()
```

```python
natural_gas_consmp_df.columns
```

```python
# Removing the unwanted columns of natural_gas_consmp_df into and storing it into new dataframe
cleaned_natural_gas_consmp_df = natural_gas_consmp_df[['year','month','duoarea','product-name','process-name','value','units']]
cleaned_natural_gas_consmp_df.head()
```

```python
#Removing S letter from our duoarea column values to get the state abbrevations 
cleaned_natural_gas_consmp_df.loc[:, 'duoarea'] = cleaned_natural_gas_consmp_df['duoarea'].str.lstrip('S')

cleaned_natural_gas_consmp_df.head()
```

```python
#Renaming the duoarea column name to state 
cleaned_natural_gas_consmp_df = cleaned_natural_gas_consmp_df.rename(columns={'duoarea':'state'})

cleaned_natural_gas_consmp_df.head()
```

```python
cleaned_aqi_df = cleaned_aqi_df.rename(columns={'state_id':'state'})

cleaned_aqi_df.head()
```
