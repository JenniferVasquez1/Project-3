```python
import sqlite3
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Category10  # Using Category10 for a set of colors
from bokeh.transform import factor_cmap

# Ensure Bokeh plots are displayed in the notebook
output_notebook()

# Connect to SQLite database
conn = sqlite3.connect('new_aqi_ngsDB.sqlite')

# Write an SQL query
query = "SELECT AQI, [Commercial Consumption], month, state_name FROM combined_df"

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(query, conn)

# Convert 'month' column to datetime format if it's not already
df['month'] = pd.to_datetime(df['month'])

# Extract month names for season classification
df['month_name'] = df['month'].dt.month

# Define a function to classify months into seasons
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Apply the function to create a 'season' column
df['season'] = df['month_name'].apply(get_season)

# Check the unique seasons in the DataFrame
print(df['season'].unique())

# Define colors for each season
season_colors = {'Winter': 'blue', 'Spring': 'green', 'Summer': 'orange', 'Fall': 'brown'}

print(df)

# Create a ColumnDataSource for Bokeh
source = ColumnDataSource(df)

# Create the correlation scatter plot
p = figure(width=800, height=400, title="Correlation between AQI and Commercial Consumption", tools="pan,wheel_zoom,box_zoom,reset,hover")

# Define a color map for seasons
color_map = factor_cmap('season', palette=list(season_colors.values()), factors=list(season_colors.keys()))

# Add a scatter renderer with size, color, and alpha
p.scatter(x='Commercial Consumption', y='AQI', source=source, marker="square", size=10, color=color_map, alpha=0.7)

# Set axis labels
p.xaxis.axis_label = 'Commercial Consumption'
p.yaxis.axis_label = 'AQI'

# Add hover tool to make the plot interactive
hover = p.select(dict(type=HoverTool))
hover.tooltips = [("Month", "@month{%F}"), ("State", "@state_name"), ("AQI", "@AQI"), ("Consumption", "@{Commercial Consumption}"), ("Season", "@season")]
hover.formatters = {"@month": "datetime"}

#Show the plot
show(p)
```
