```python
AQI_df = pd.read_csv('../Resources/AQI.csv', index_col=0)
```

```python
AQI_df.head()
```

```python
drops=['Unnamed: 0','CBSA Code','Defining Parameter','timezone','Number of Sites Reporting', 'city']
```

```python
AQI_df.drop(drops,axis=1,inplace=True)
```

```python
AQI_df.head()
```

```python
AQI_df.rename({'city_ascii': 'city'}, axis=1, inplace=True)
```

```python
AQI_df.groupby(['city', 'state_id']).size()
```

```python
AQI_df = AQI_df.reset_index()
```

```python
AQI_df.Date = AQI_df.Date.str.replace('2021-', '').replace('-.+', '', regex=True)
```

```python
AQI_df
```

```python
AQI_df.rename({'Date': 'month'}, axis=1, inplace=True)
```

```python
AQI_df
```

```python
grp = AQI_df.groupby(['state_id', 'month'])
```

```python
new_AQI_df = grp[['AQI', 'lat', 'lng', 'population', 'density']].mean()
new_AQI_df = new_AQI_df.reset_index()
new_AQI_df.head()
```

```python
new_AQI_df[['AQI', 'population', 'density']] = new_AQI_df[['AQI', 'population', 'density']].astype('int')
new_AQI_df.head()
```

```python
new_AQI_df.lat = new_AQI_df.lat.map(lambda x: round(x, 4))
new_AQI_df.lng = new_AQI_df.lng.map(lambda x: round(x, 4))
```

```python
new_AQI_df.head()
```

```python
new_AQI_df['Category'] = pd.cut(new_AQI_df.AQI, [0,50,100,150,200,300,666], 
       labels=['Good','Moderate','Unhealthy for Sensitive Groups','Unhealthy','Very Unhealthy','Hazardous'])
```

```python
new_AQI_df[:5]
```

```python
new_AQI_df.to_csv('../Resources/new_AQI.csv')
```
